Exercicio 01


```python
import matplotlib.pyplot as plt
```

```python
ano = [2020, 2021, 2022, 2023]
aprovados = [40, 55, 60, 70]
plt.plot(ano, aprovados, marker="o")
plt.grid(True)
plt.title("Aprovados até 2023")
plt.ylabel("Alunos aprovados")
plt.xlabel("Ano")
plt.xticks(ano)
plt.yticks(aprovados)
```
